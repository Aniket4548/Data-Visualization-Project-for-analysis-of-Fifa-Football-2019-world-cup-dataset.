```python
plt.hist(df['Overall']) #histogram are approximations to distributions generated through value based on a set
# x axis = range, y axis = frequency
```

```python
plt.hist(df['Overall'], edgecolor = 'black' , bins = 15)  #create histogram and specifying bin size and border color
plt.xlabel ('Overall')  #giving label to x axis
plt.ylabel ('Frequency')  #giving label to y axis
plt.title ('Histogram of Overall Player Ratings in FIFA19')  #giving histogram title
plt.show()  #display graph
```

```python
width = 5
plt.hist(df['Overall'], edgecolor = 'black' , bins = np.arange(min(df['Overall']),max(df['Overall']) + width ,width))
#create histogram and specifying bin size and border color
plt.xlabel ('Overall')  #giving label to x axis
plt.ylabel ('Frequency')  #giving label to y axis
plt.title ('Histogram of Overall Player Ratings in FIFA 19')  #giving histogram title
plt.show()  #display graph
```

```python
#no.of bins = [log2n + 1]
#optimal no. of bins = [log2n + 1] = log2 [(18207)+1] = 15 approx
plt.hist(df['Overall'], edgecolor = 'White' , facecolor = 'red' , bins = 15 , range = [50,90])
plt.xlabel ('Overall')  #giving label to x axis
plt.ylabel ('Frequency')  #giving label to y axis
plt.title ('Histogram of Overall Player Ratings in FIFA 19')  #giving histogram title
plt.show()  #display graph
```

```python
#Bar chart
#analysing categorical data
from collections import Counter  #helps in counting the value of each value
print(Counter(df['Nationality']))
#you will get a dictionary as an output
```

```python
#Filter out the most common nationality values
print (dict(Counter(df['Nationality']).most_common(15)))
```

```python
#To filter out most uncommon nationaality values
least_common = Counter(df['Nationality']).most_common()[::-1]
print(least_common[0:10])
```

```python
print(least_common[0:30])
```

```python
nationality_dict = dict(Counter(df['Nationality']).most_common(10))
plt.bar(nationality_dict.keys(),nationality_dict.values())
plt.xlabel('Nationality')
plt.ylabel('Frequency')
plt.title('Bar Plot of Ten Most CommonCountries')
```

```python
nationality_dict = dict(Counter(df['Nationality']).most_common(10))
plt.bar(nationality_dict.keys(),nationality_dict.values())
plt.xlabel('Nationality')
plt.ylabel('Frequency')
plt.title('Bar Plot of Ten Most Common Nationality')
plt.xticks(rotation=90)  #rotates the x axis label by 90 degree
plt.show()
```

```python
least_common = Counter(df['Nationality']).most_common()[::-1]
least_common = dict(least_common[:30])
plt.bar(least_common.keys(), least_common.values())
plt.xlabel('Nationality')
plt.ylabel('Frequency')
plt.title('Bar Plot of Ten Least Common Nationality')
plt.xticks(rotation=90)  #rotates the x axis label by 90 degree
plt.show()
```

```python
#Pie Chart
#visualize proportion in your data 
df.loc[df.Nationality == 'England', 'Nationality_2'] = 'England'
df.loc[df.Nationality == 'Germany', 'Nationality_2'] = 'Germany'
df.loc[df.Nationality == 'Spain', 'Nationality_2'] = 'Spain'
df.loc[df.Nationality == 'Argentina', 'Nationality_2'] = 'Argentina'
df.loc[df.Nationality == 'Brazil', 'Nationality_2'] = 'Brazil'
df.loc[df.Nationality == 'Italy', 'Nationality_2'] = 'Italy'
df.loc[~ df.Nationality.isin(['England','Germany','Spain','Argentina','Brazil','Italy']) , 'Nationality_2'] = 'Other'
#create a dictionary that contains the proportion value of these:
proportion = dict(Counter(df['Nationality_2']))
for key, values in proportion.items() : 
    proportion[key] = (values)/len(df)*100
print(proportion)
```

```python
plt.pie(proportion.values(),labels = proportion.keys(), autopct = '%1.2f%%')
plt.figure(figsize = (20,15))
plt.show()
```

```python
explode = (0,0,0,0,0,0.3,0)
plt.pie(proportion.values(),labels = proportion.keys(), autopct = '%1.2f%%' , explode=explode)
plt.figure(figsize = (20,15))
plt.show()
```

```python
import seaborn as sns
```

```python
sns.distplot(df['Overall'])  #dist plot shows the distribution and shows us the nature of the plot wether the distribution is normal right skewed or right skewed
plt.xlabel('Overall')
plt.ylabel('Frequency')
plt.title('Histogram of Overall Rating')
plt.grid()
plt.show()
```

```python
sns.scatterplot(df['Overall'], df['Wage'])
plt.xlabel('Overall')
plt.ylabel('Frequency')
plt.title('Overall vs Wage')
plt.grid()
plt.show()
```

```python
#heatmap
corr = df[['Overall','Age','Wage','Skill Moves','Potential']].corr()
sns.heatmap(corr, annot= True)  #we set annot for annotate to true to see the corellation values
plt.title('Heatmap of Overall ,Age, Wage_Euro and Skill Moves')
plt.show()
```
